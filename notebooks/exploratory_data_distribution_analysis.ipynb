```python
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

## Load a dataset

Replace the sample path with the CSV you want to analyze.

```python
DATA_PATH = Path('../data/dataset.csv')
if not DATA_PATH.exists():
    raise FileNotFoundError(f'Add a CSV file at {DATA_PATH.resolve()}')
df = pd.read_csv(DATA_PATH)
df.head()
```

## Dataset overview

```python
print('Shape:', df.shape)
display(df.dtypes.to_frame('dtype'))
display(df.isna().sum().to_frame('missing_values'))
display(df.describe(include='all').transpose())
```

## Numerical distributions

```python
numeric_columns = df.select_dtypes(include=np.number).columns.tolist()
if not numeric_columns:
    raise ValueError('The dataset contains no numerical columns.')

for column in numeric_columns:
    values = df[column].dropna()
    plt.figure(figsize=(9, 5))
    plt.hist(values, bins=30, edgecolor='black')
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

## Box plots and outlier counts

```python
outlier_summary = []
for column in numeric_columns:
    values = df[column].dropna()
    q1, q3 = values.quantile([0.25, 0.75])
    iqr = q3 - q1
    lower, upper = q1 - 1.5 * iqr, q3 + 1.5 * iqr
    outliers = values[(values < lower) | (values > upper)]
    outlier_summary.append({'column': column, 'outliers': len(outliers), 'lower_bound': lower, 'upper_bound': upper})

    plt.figure(figsize=(9, 3))
    plt.boxplot(values, vert=False)
    plt.title(f'Box plot of {column}')
    plt.xlabel(column)
    plt.show()

pd.DataFrame(outlier_summary)
```

## Correlation analysis

```python
correlation = df[numeric_columns].corr()
plt.figure(figsize=(10, 7))
sns.heatmap(correlation, annot=True, fmt='.2f')
plt.title('Correlation matrix')
plt.show()
```

## Distribution diagnostics

```python
diagnostics = []
for column in numeric_columns:
    values = df[column].dropna()
    diagnostics.append({
        'column': column,
        'count': len(values),
        'mean': values.mean(),
        'median': values.median(),
        'std': values.std(),
        'skewness': stats.skew(values) if len(values) > 2 else np.nan,
        'kurtosis': stats.kurtosis(values) if len(values) > 3 else np.nan,
    })
pd.DataFrame(diagnostics)
```
